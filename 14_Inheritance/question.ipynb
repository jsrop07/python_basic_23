```python
# 클래스 문제6
def nested_function():
    status = '작성중'
    def refunction():
        nonlocal status
        status = "완료"
    print(status)

    refunction()
    print(status)
```

```python
class Mentor:
    def guide(self):
        return "d이직 준비스"
class Lecturer:
    def teach(self):
        return "이직에 필요한 기술을 가르쳐드립니다."
class Coach(Mentor,Lecturer):
    def support(self):
        return "여러방면으로 지원을 아낌"

sw = Coach()
sw.support()
```

```python
class Mentor:
    def __init__(self,field):
        self.field = field
    def guide(self):
        return f"{self.field}반환"
class Lecturer:
    def __init__(self,experience):
        self.experience = experience
    def teach(self):
        return f"{self.experience}경력스"

class Candidate(Mentor, Lecturer):
    def __init__(self,field,experience):
        Mentor.__init__(self,field)
        Lecturer.__init__(self,experience)
    def guide(self):
        return f"저는 {self.field}"
sw1 = Mentor('D')
print(sw1.guide())
sw = Candidate('D','S')
sw.guide()
        
```
