```python
customers = '토끼', '곰', '여우', '호랑이', '늑대'
count =0
for i in customers:
    print(f"어서오세요 {i}님")
    count +=1
    if(count==3):
        print("end")
        break
```

```python
customers = ['토끼', '곰', '여우', '호랑이', '늑대']
for i in range(len(customers)):
    if i == 3:
        print('죙송')
        break
    print(f"어서오세요 {customers[i]}님")
```

```python
customers = ['토끼', '곰', '여우', '호랑이', '늑대']
for idx, customers in enumerate(customers):
    if(idx ==3 ):
        print('죙송')
        break
    print(f"어서오세요 {customers}님")
```

```python
orders = ['아메리카노', '카페라떼', '핫초코', '모르겠어요', '에스프레소']
for i in orders:
    if(i == '핫초코'):
        print("주문에 문제가 있음")
        break
    print(f"{i}주문 확인")
```
