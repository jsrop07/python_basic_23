```python
for i in range(3):
    print(f"{i}번째 시작")
    for j in range(3):
        print(i,j)
    print(f"{i,j} 구분")
```

```python
count =0
for i in range(3):
    for j in range(3):
        for k in range(3):

            count +=1
print(count)
```

```python
for x in range(3):
    print("front",x)
    if x == 1:
        print("if",x)
        print("continue 등장")
        continue
    print("back", x)
```

```python
for i in range(3):
    for j in range(3):
        print(i,j)
        break
# break문을 만나게 되서 i가 0일때 종료 후i가 1일때 또 만나서 종료가 되기에 
# j=0으로 계속 남을 수 밖에 없음 즉, 하나의 반복문만 돌게된다
```
