```python
class Person:
    national = 'korea' # 속성
    language = 'Korean' # 속성

    def __init__(self, name, age): # 객체를 정의하는 속성으로 쓰겠다
        self.name =name # 생성자
        self.age = age 

    def greeting(self): # 메서드
        return 'hello this is python'
    
    def information(self): # 메서드
        return "I'm form" + self.national + " and I use" +self.language 
    
    def favorate(self, color):
        return "I love " + color
```

```python
# 파이썬 파일 namespace.py
# 전역 변수
variable = 'gloval variable'
print('전역', variable)

def outer_function():
    # 외부 변수
    variable = "outer variable"
    print('지역(외부함수)',variable)

    def inner_function():
        variable = "inner variable"
        print('지역(내부함수): ', variable)

    inner_function()
outer_function()

```

```python
g_variable = 'global variable'

def modify_global():
    global g_variable
    g_variable = 'global modified in function'

print(g_variable)
modify_global()
print(g_variable)
```

```python
def outer_function():
    variable = "outer variable"

    def inner_function():
        nonlocal variable #중첩 함수에서 한 단계 바깥의 함수 변수에 접근
        variable = "outer modified in inner function"

    print(variable)
    inner_function()
    print(variable)

outer_function()    # outer variable
                    # outer modified in inner function
```

```python
class Character:
    def __init__(self, nickname, type):
        self.nickname = nickname      # Public attribute
        self.__type = type            # Private attribute

    def get_type(self):
        return self.__type

    def set_type(self, type):
        character_types = ['전사', '법사', '치유사']
        for given_type in character_types:
            if(type == given_type):
                self.__type = type
                return
        
        raise ValueError("잘못된 캐릭터 타입을 선택하였습니다.")

my_char = Character('모와이','전사')
print(my_char.get_type())


my_char2 =Character('카피','힐러')
print(my_char2.nickname)
print(my_char2.set_type('법사'))

            

        
```
