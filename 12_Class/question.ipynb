```python
class Mowai:
    species = '다람쥐'
    def __init__(self, name, current_job): # 초기화를 해야지 인스턴스를 새로 만들때 그 값에 해당하는 매개변수들을 출력할수 있다
        self.name = name
        self.current_job = current_job

    def introduce(self): #self라는 매개변수는 무조건 필수
        return "hello ma name is " + self.name + " My job is " + self.current_job  
```

```python
mowai1 = Mowai('석원','백수') # 인스턴스를 생성하는 구문(고유)
mowai2 = Mowai('원석','학생')

print(mowai1.species, mowai2.species) 
print(mowai1.introduce())
print(mowai2.introduce())
```

```python
# 2
class Character:
    skills =[] # 모든 인스턴스가 같이 공유 그렇기에 초기화가 필수

    def __init__(self):
        self.skills = []
    
    def add_skill(self, skill):
        self.skills.append(skill)
        

```

```python
char1 = Character()
char1.add_skill('밥 먹음')
print(char1.skills)
```

```python
char2 = Character()
char2.add_skill('밥 안 먹음')
print(char2.skills)
```

```python
# 4 
tool = '연필'
def namespace_test():
    local_tool = '키보드'
    global tool
    tool = '펜'
    print(local_tool)
    print(tool)

print(tool)

namespace_test()
```

```python
class Toolbox():
    tools = ['노트북']

    def __init__(self): #자기자신이 가지고있는 tools = self
        self.tools = []
    
    def add_tool(self,too):
        self.tools.extend(too)

tool2 = Toolbox()
tool2.add_tool(['빠라밤','asd'])
print(tool2.tools)
```

```python
class Toolbox():    
    tools = ['노트북', '책상']

    def __init__(self):
        self.tools = []

    def add_tool(self, tool):
        self.tools.append(tool)

    def remove_tool(self, tool):
        self.tools.remove(tool)

temp5 = Toolbox()

temp5.add_tool("테스트3")
#temp5.remove_tool("테스트3")
print(temp5.tools)
```

```python
def nested_function():
    status = '작성중'

    def refucntion():
        global status
        status = '완료'
    print(status)
    refucntion()
    print(status)

nested_function()
```
