```python
# 파이썬 파일 namespace.py
# 전역 변수
variable = 'gloval variable'
print('전역', variable)

def outer_function():
    # 외부 변수
    variable = "outer variable"
    print('지역(외부함수)',variable)

    def inner_function():
        variable = "inner variable"
        print('지역(내부함수): ', variable)

    inner_function()
outer_function()

```

```python
class TestClass:
    # 클래스 네임스페이스
    variable = "class variable"

    def __init__(self, value):
        self.variable = value
    
    def class_function(self):
        variable = "local variable"
        print('클래스 지역: ', variable)

obj =TestClass('instance variable')

print('인스턴스: ', obj.variable)
obj.class_function()
```

```python
g_variable = 'global variable'

def modify_global():
    global g_variable
    g_variable = 'global modified in function'

print(g_variable)
modify_global()
print(g_variable)
```
