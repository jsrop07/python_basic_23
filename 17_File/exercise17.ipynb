```python
class Squirrel:
    species = '다람쥐'
    def __init__(self,name,current_job):
        self.name = name
        self.current_job = current_job
    def introduce(self):
        return f'{self.name},{self.current_job}'

sw=Squirrel('람쥐','썬더')
sw.introduce()
```

```python
class Character:
    skills = []
    def __init__(self,skills):
        self.skills = skills
    def add_skill(self):
        return f"{self.skills}"

sw2 = Character('skills')
sw2.add_skill()
    
```

```python
tool ='연필'

def namespace_test():
    local_tool = '키보드'
    global tool
    tool = '펜'
    print(local_tool)
namespace_test()
print(tool)
```

```python
class Toolbox:
    tools =['노트북', '책상']

    def __init__(self,tools):
        self.tools = tools 
    def add_tool(self):
        Toolbox.tools.append(tools)

sw3 = Toolbox('키키키')
sw3.add_tool()

print(sw3.tools)
```

```python
def nested_function():
    status = '작성중'
    def inner():
        nonlocal status
        status = '완료'
    print(inner())
print(inner())
```
