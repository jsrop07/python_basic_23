```python
# csv형태로 저장
import csv

data = [
    ["회사명", "직무",'결과'],
    ["회사A", "데이터 분석가",'합격'],
    ["회사B", "소프트웨어 엔지니어",'불합격'],
    ["회사C", "제품 관리자",'진행중'],
]

with open("application.csv","w",newline ="",encoding='utf-8') as f:
    writer = csv.writer(f) # csv 작성도구 
    writer.writerows(data) # 여러행을 한꺼번에 작성

with open("application.csv","r",newline ="",encoding='utf-8') as f:
    reader = csv.reader(f) #csv reader
    for row in reader:
        print(",".join(row))

```

```python
# 딕셔너리 형태로 저장
import csv

data = [
    {"회사명"  : "회사1", "직무":"데이터 분석가",'결과': "합격"},
    {"회사명"  : "회사2", "직무":"빅데이터",'결과': "합격"},
    {"회사명"  : "회사3", "직무":"제품 관리자",'결과': "불합"},
]

with open("application.csv",'w',encoding='utf-8') as f:
    fieldnames = ["회사명", "직무","결과"]
    writer = csv.DictWriter(f,fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)

with open("application.csv",'r',encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row)
```

```python
# 문제 4
# json없이 json출력
with open('favorate_foods.json','w', encoding='utf-8') as json_file:
    json_file.write('{"name" : "카피바라","favorate_foods":["과자","고기"]}')

with open('favorate_foods.json','r',encoding='utf-8') as json_file:
    json_text = json_file.read()
    # print(" ".join(json_text['favorate_foods']))
    print((json_text.split("[")[1].split("]")[0]))
```

```python
# 문제 3
with open('./system.log','w',encoding='utf-8') as ff:
    log_content = ['[INFO] 2024-01-01 12:00:00: 시스템 시작\n[WARN] 2024-01-01 12:05:00: 메모리 사용량 증가\n[ERROR] 2024-01-01 12:10:00: 프로그램 종료']
    ff.writelines(log_content)

with open('system.log','r',encoding='utf-8') as ff:
    for line in ff:
        if "WARN" in line or 'ERROR' in line:
            print(line)
```

```python
# 문제 4
import json

my_dict = {"name": "다람쥐","favorite_foods": ["도토리", "호두", "밤"]}
with open('./foods.json','w', encoding='utf-8') as fff:
    json.dump(my_dict,fff,ensure_ascii=False)
with open('foods.json','r',encoding='utf-8') as fff:
    contents =fff.read()
    cc =contents.split("[")[1].split("]")[0]
    print(cc)
```
