```python
javaTeam = {'Squirrel','Gorilla', 'Tiger', 'Sheep', 'monkey', 'wolf'}
pythonTeam = {'Pig', 'Squirrel', 'Bear', 'Gorilla'}

ohgiraffers = javaTeam.union(pythonTeam)
print(javaTeam | pythonTeam)    # {'wolf', 'Sheep', 'Tiger', 'Gorilla', 'monkey', 'Squirrel', 'Pig', 'Bear'}
print(ohgiraffers)              # {'wolf', 'Sheep', 'Tiger', 'Gorilla', 'monkey', 'Squirrel', 'Pig', 'Bear'}
```

```python
new_mowai =  javaTeam | pythonTeam # 합집합 
print(new_mowai)
```

```python
print(javaTeam & pythonTeam)                # {'Squirrel', 'Gorilla'}
print(javaTeam.intersection(pythonTeam))    # {'Squirrel', 'Gorilla'}
```

```python
# difference
print(javaTeam - pythonTeam)              # {'monkey', 'wolf', 'Sheep', 'Tiger'}
print(javaTeam.difference(pythonTeam))    # {'monkey', 'wolf', 'Sheep', 'Tiger'}
# 좌항을 기준으로 우항의 차집합 변환 
```

```python
# 공통 차집합 (교집합이 없는 각각의 부분 )
print(javaTeam.symmetric_difference(pythonTeam))
print(javaTeam ^ pythonTeam)
```

```python
# 1.shallow copy
a = {'a': 1, 'b': 2, 'c': 3}
b = a # 사전 레퍼런스의 복사 (공유됨)
print(a) 
print(b)

```

```python
# 참조하는 공간의 주소값을 복사 주소값을 공유
# a라는 변수의 공간의 주소값을 같이 쓰기에 공유가 된다(내부값)
a['a'] = 100
print(a)
print(b)
```

```python
# 2.deep copy
c = {'a': 1, 'b': 2, 'c': 3}
d = c.copy()
print(c)
print(d)

```

```python

c['a'] = 300
print(c)
print(d) # 변경되지 않음.
```

```python
c = {'a' : [1,2,3]}
d= c.copy()

c['a'][0] = 999
print(c)
print(d)
# 하나의 같은 공간을 바라보게 된다?
# .copy를 했으면 주소값을 참조
# 그냥 카피하면 아예 다른 공간을 참조
```

```python
# value에 list가 있다면 copy메소드를 사용해도 list는 얕은 복사되어 변수 간 공유된다.
a = {'a': [1,2,3], 'b':100}
b = a.copy() #=> 얕은 복사 : 같은 주소
# deepcopy() => 깊은 복사 : 다른 주소
print(a) #{'a': [1, 2, 3], 'b': 100}
print(b) #{'a': [1, 2, 3], 'b': 100}

a['b'] = 200
a['a'][0] = 999
print(a) #{'a': [999, 2, 3], 'b': 200}
print(b) #{'a': [999, 2, 3], 'b': 100}

# value에 dict가 있다면
a = {'user':{'name':'김동현','age':20},'amount':123}
b = a.copy()
print(a)
print(b)

a['user']['name'] = '김동팔'
print(a)
print(b)
```

```python
import copy 

original = {
    'name' : 'capiara',
    'score' : [100,90,80]
}

deep = copy.deepcopy(original) # 값만 복사될뿐 완전 다른 주소

original['score'][0] = 999

print('original : ', original)
print('deep : ', deep)
```

```python
# update 메소드를 통한 두 dict의 병합
# 동일한 키값이 있다면, 인수로 주어진 dict의 아이템으로 덮어쓰기 된다.
dic1 = {'a':10, 'b':20}
dic2 = {'c':100, 'd':300}
dic1.update(dic2)

print(dic1) #{'a': 10, 'b': 20, 'c': 100, 'd': 300}
print(dic2) #{'c': 100, 'd': 300}
```

```python
# dict의 정렬 : sorted
numbers = {'first': 1, 'second': 2, 'third': 3, 'Fourth': 4}
numbers['hundred'] = 100
numbers['fifth'] = 5

print(list(numbers)) #['first', 'second', 'third', 'Fourth', 'hundred', 'fifth'] 형변환(키값만 남음)

# key값 정렬
print(sorted(numbers)) #['Fourth', 'fifth', 'first', 'hundred', 'second', 'third']

# value값 정렬
print(sorted(numbers.values())) #[1, 2, 3, 4, 5, 100]

# key정렬 (value값 정렬에 따라)
print(sorted(numbers, key=numbers.__getitem__))
numbers
```
