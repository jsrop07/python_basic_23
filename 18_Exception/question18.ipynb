```python
try:
    age = int(input('입력'))
except ValueError as e:
    print('숫자를 입력',e)
else:
    if age>=19:
        print('성인')
    else :
        print('미성년자')
```

```python
try:
    f = int(input('입력'))
    f2 = int(input('입력2'))
    f/f2
except ValueError as e:
    print(f' 숫자 입력{e}')
except ZeroDivisionError as e:
    print("0으로 안됨",e)
    


```

```python
try:
    with open('dairy.txt','r',encoding='utf-8') as f:
        print(f.read())
except FileNotFoundError:
    print('파일 없음')
finally:
    if f:
        f.close()
    print('파일 자원 정리 완료')
    
```

```python
import csv
with open('appication.csv','w',encoding='utf-8') as f:
    f.write('회사명,직무,결과\n \
            회사A,데이터 분석가,합격\n \
            회사B,소프트웨어 엔지니어,불합격\n \
            회사C,제품 관리자,진행 중')

with open('appication.csv','r',encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        print(",".join(row))
    
```
