```python
from collections import deque # colletions란 함수들을 하나로 모아놓은 라이브러리(그 중에 deque라는 자료구조만 쓰겠다)
dq = deque()
dq.append("a")
dq.append("b")
dq.append("c")
print(dq)

```

```python
dq.appendleft('x')
dq
```

```python
value = dq.pop()
value
```

```python
left_value = dq.popleft()
print(left_value)
```

```python
# 리스트
location = ['서울특별시', '부산광역시', '인천광역시', '광주광역시', '울산광역시']
print('광주광역시' in location)                   # True
print(['광주광역시', '부산광역시'] in location)    # False 하나의 요소로 들어가있지 않기 때문에 false 

print(['서울특별시', '부산광역시'] in location)    # False
```
