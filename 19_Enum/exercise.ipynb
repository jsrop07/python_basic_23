```python
from enum import Enum
# name = value 지정 필수
class Gender(Enum):
    M = '남'
    F = '여'

print(Gender.M)
print(Gender.F)
```

```python
# name
print(Gender.M.name)
print(Gender.F.name)

# value
print(Gender.M.value)
print(Gender.F.value)
```

```python
a = Gender.M
b = Gender.M
print(a)

print(b)
```

```python
print(a==b)
print(a is b)
```

```python
for gender in Gender:
    print(gender)
    
```

```python
Color = Enum('Color', ['RED','apple', 'GREEN', 'BLUE'])
for color in Color:
    print(color.name, color.value)
```

```python
from enum import auto

class Lang(Enum):
    # auto()로 지정되는 값을 override 가능
    # member선언보다 먼저 작성할 것
    def _generate_next_value_(name, start, count, last_value):
        # return name # name, value 동일
        return start # 첫번째 요소값 - 동일한 value가 지정되는 경우 선언무시된다?
        # return count # 0 1 2 인덱스 지정
        # return last_value # 마지막에 지정된 value

    C = auto()        # 1
    Java = auto()     # 2
    Python = auto()   # 3

for lang in Lang:
    print(lang.name, lang.value)
```

```python
# 정해진 값만 허용하고 싶을때
class Status(Enum):
    READY = 'READY'
    RUNNING = 'RUNNING'
    DONE = 'DONE'

for ss in Status:
    print(ss.name, ss.value)
```
