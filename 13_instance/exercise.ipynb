```python
class Person:
    national = 'korea' # 속성
    language = 'Korean' # 속성

    def __init__(self, name, age): # 객체를 정의하는 속성으로 쓰겠다
        self.name =name # 생성자
        self.age = age 

    def greeting(self): # 메서드
        return 'hello this is python'
    
    def information(self): # 메서드
        return "I'm form" + self.national + " and I use" +self.language 
    
    def favorate(self, color):
        return "I love " + color
```

```python
mowai = Person('모와이',900)
mowai2 = Person('고릴라',20)
print(mowai.national)
print(mowai.greeting)
print(mowai2.information)
print(mowai2.favorate('blue'))
```

```python
class Character:    
    skills = []
    
    def __init__(self, nickname, type):
        self.nickname = nickname
        self.type = type
        
    def add_skill(self, skill):
        self.skills.append(skill)
    
    def show_skill(self):
        return self.type + ' ' + self.nickname + '님의 보유 스킬 ' + str(self.skills)

my_character = Character('산골짜기다람쥐', '전사')
your_character = Character('흑염룡의고릴라', '법사')
my_character.add_skill('곤봉 휘두르기')
your_character.add_skill('마법진 그려서 날려버리기')

print(your_character.show_skill())    # 전사 산골짜기다람쥐님의 보유 스킬 ['곤봉 휘두르기', '마법진 그려서 날려버리기']
```
